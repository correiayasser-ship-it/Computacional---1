```python
# Aluno: Yasser Correia De Oliveira
# Professores: Mauricio Gabriel Lacerda Mafra / Renato Ramos Da Silva Dantas
# Dicisplina: Geofsica Computacional 1
# Atividade de Sistemas Lineares (Matrizes e Vetores)

import numpy as np
import matplotlib.pyplot as plt

def Regressao(a = -1, b = 2, c = 0, d = 0, e = 0, f = 3, grau = None):

    x = np.linspace(-10, 10, 100)

    coef = np.array([a, b, c, d, e, f])

    if grau is None:
        idx = np.where(coef != 0)[0]
        grau = len(coef) - np.argmax(coef != 0) - 1


    y = np.polyval(coef[:grau+1], x)
    X = np.vstack([x**i for i in range(grau, -1, -1)]).T

    theta = np.linalg.inv(X.T @ X) @ (X.T @ y)

    return x, y, theta

x, y, theta = Regressao()
coeficientes = len(theta)
X = np.vstack([x**i for i in range(coeficientes - 1, -1, -1)]).T
y_reg = X @ theta

plt.figure(figsize=(8,5))
plt.plot(x, y, label=r"Funo original: $-x^5 + 2x^4 + 3$")
plt.plot(x, y_reg, '--', label="Regresso linear/polinomial")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Funo vs Regresso Linear/Polinomial")
plt.grid(True)
plt.legend()
plt.show()
```
