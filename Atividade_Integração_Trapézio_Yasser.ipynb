```python
# Aluno: Yasser Correia De Oliveira
# Professores: Mauricio Gabriel Lacerda Mafra / Renato Ramos Da Silva Dantas
# Dicisplina: Geofísica Computacional 1
# Atividade de Integração - Trapézio

import matplotlib.pyplot as plt

def Integracao_Trapezio(delta_x = 0.01, x_i = -2, x_f = 4, a = -1, b = 2, c = 0, d = 0, e = 0, f = 3, plot = False):
    integracao = []
    for i in range (int((x_f - x_i)/delta_x) + 1):
       integracao.append(x_i + i*delta_x)

    fxi = []
    for i in range(len(integracao)):
        x = integracao[i]
        f_x_i = a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f
        fxi.append(f_x_i)

    Integral_trapezio = (delta_x/2) * (fxi[0] + 2*sum(fxi[1:-1]) + fxi[-1])

    integral_xi = (a/6)*x_i**6 + (b/5)*x_i**5 + (c/4)*x_i**4 + (d/3)*x_i**3 + (e/2)*x_i**2 + f*x_i
    integral_xf = (a/6)*x_f**6 + (b/5)*x_f**5 + (c/4)*x_f**4 + (d/3)*x_f**3 + (e/2)*x_f**2 + f*x_f
    integral_analitica = integral_xf - integral_xi

    dicionario = {
      'Integração': integracao,
      'f(x)': fxi,
      'Integral Numérica (Trapézio)': Integral_trapezio,
      'Integral Analítica': integral_analitica
      }

    if plot == True:
      coef = [a, b, c, d, e]
      pot = [5, 4, 3, 2, 1]

      partes = []
      for A, P in zip(coef, pot):
          if A == 0:
              continue
          if A == 1:
              partes.append(f"x^{P}")
          elif A == -1:
              partes.append(f"-x^{P}")
          else:
              partes.append(f"{A}x^{P}")

      if f != 0:
          partes.append(f"{f}")

      expressao_fx = " + ".join(partes).replace("+ -", "- ")

      plt.figure(figsize=(10, 6))
      plt.plot(integracao, fxi, 'g',
               label = rf'$f(x) = {expressao_fx}$')

      for i in range(len(integracao)):
          x0 = integracao[i]
          plt.bar(x0, fxi[i], width=delta_x, align='edge', color='b', edgecolor='black', alpha=0.7)

      plt.text(0.57, 0.09,
               rf'$\int_{{{x_i}}}^{{{x_f}}} ({expressao_fx})\, dx = {Integral_trapezio:.2f}$',
               transform=plt.gca().transAxes,
               fontsize=12,
               verticalalignment='top',
               bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.7))

      plt.text(0.01, 0.02,
               rf'Integral Analítica = {integral_analitica:.2f}',
               transform=plt.gca().transAxes,
               fontsize=12,
               bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.7))

      plt.title('Integração Númerica - Regra do Trapézio')
      plt.xlabel(r'$x$')
      plt.ylabel(r'$f(x)$')
      plt.legend()
      plt.savefig('Integração_Trapézio.pdf')
      plt.show()

    return dicionario
```

```python
Integracao_Trapezio(delta_x = 0.01, x_i = -2, x_f = 4, a = -1, b = 2, c = 0, d = 0, e = 0, f = 3, plot = True)
```
