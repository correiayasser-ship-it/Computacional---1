```python
# Aluno: Yasser Correia De Oliveira
# Professores: Mauricio Gabriel Lacerda Mafra / Renato Ramos Da Silva Dantas
# Dicisplina: Geofísica Computacional 1
# Atividade de Derivação

import numpy as np
import matplotlib.pyplot as plt

def Derivada(delta_x = 0.001 , x = 45, funcao = np.sin, plot = True):

    x_rad = x*np.pi/180
    delta_rad = delta_x*np.pi/180

    derivada1 = (funcao(x_rad + delta_rad) - funcao(x_rad - delta_rad)) / (2 * delta_rad)
    derivada2 = (funcao(x_rad + delta_rad) + funcao(x_rad - delta_rad) - 2 * funcao(x_rad)) / (delta_rad**2)

    dicionario = {
        'f(x)': funcao(x_rad),
        'Derivada 1ª': derivada1,
        'Derivada 2ª': derivada2
    }

    if not plot:
        return dicionario

    xs = np.linspace(0, 360, 400)

    f_vals  = []
    d1_vals = []
    d2_vals = []

    for ang in xs:
        ang_r = ang*np.pi/180
        f_vals.append(funcao(ang_r))
        d1_vals.append((funcao(ang_r + delta_rad) - funcao(ang_r - delta_rad)) / (2 * delta_rad))
        d2_vals.append((funcao(ang_r + delta_rad) + funcao(ang_r - delta_rad) - 2 * funcao(ang_r)) / (delta_rad**2))

    plt.figure(figsize=(12, 6))
    plt.plot(xs, f_vals, label='f(x)')
    plt.plot(xs, d1_vals, label='1ª derivada (numérica)')
    plt.plot(xs, d2_vals, label='2ª derivada (numérica)')

    texto = (
        f"x = {x}°\n"
        f"f(x) = {dicionario['f(x)']:.5f}\n"
        f"1ª derivada = {dicionario['Derivada 1ª']:.5f}\n"
        f"2ª derivada = {dicionario['Derivada 2ª']:.5f}"
    )

    plt.text(
        0.01, 0.15, texto,
        transform=plt.gca().transAxes,
        verticalalignment='top',
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='black')
    )

    plt.xlabel("x (graus)")
    plt.ylabel("valor")
    plt.title("Função e derivadas numéricas")
    plt.legend()
    plt.grid(True)
    plt.show()

    return "Gráfico exibido."
```

```python
Derivada(delta_x = 0.001, x = 45, funcao = np.sin , plot = True)
```
